# Pandas Melt Function

```python
import pandas as pd
df = pd.read_csv('weather.csv')
df.head()
```

```python
df1 = pd.melt(df, id_vars=['day'])   # id_vars = id variable, x axix or column, melt works as pivot to some extend
df1 
```

```python
df1[df1['variable']=='Dhaka']  # Accessing
```

```python
df1[df1['variable']=='Sylhet']  # Accessing
```

#### Changing Column Names

```python
df1 = pd.melt(df, id_vars=['day'], var_name = 'City', value_name = 'Temperature') 
df1
```
