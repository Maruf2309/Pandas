### Creating Two Data Frame

```python
df1 = pd.DataFrame({'id':[1,2,3,5,9],
                    'Col_1':[1,2,3,4,5],
                    'Col_2':[6,7,8,9,10],
                    'Col_3':[11,12,13,14,15],
                    'Col_4':['apple','orange','banana','strawberry','resberry']
                    
                   })

df2 = pd.DataFrame({'id':[1,1,3,5],
                     'Col_A':[8,9,10,11],
                     'Col_B':[12,13,15,17],
                     'Col_4':['apple','orange','banana','lichy'],
                   })
```

### About pandas merge()

- For combining data on common columns
- Many-to-one and many-to-many joins are possible
- Side by side merge

# Inner Join / Join

- Merge will join two tables with their common elements
- Here in this two data frame matching cols are 'id' & 'Col_4'
- It will present the data frame side by side as per their common id numbers
- No match no ruturn in new data frame
- Not only ID values, rather "COLUMNS NAMES" MUST MATCHED, if COLUMNS NAME MATCHED, then it will join table as per id values
- If columns names not matched, then need to join with specific conditions

```python
inner = pd.merge(df1, df2)
inner
```

 ### Join tables with conditions - Join explicitly - Recommended Approach

- See Id number 1 is retuns twice, as its working for "One-to-Many"
- See the join 'Col_4_x' & 'Col_4_x'; it means these cols are common in both data frame
- 'Col_4_x' : is associate with left dataframe
- 'Col_4_y' : is associate with right dataframe
- Suffix (x & y) can be renamed as per requirements by using suffixes function

```python
pd.merge(df1, df2, on = 'id')   # on arguments are case sensitive 
```

```python
# this is the same as the first merge becaus we are specifing all the columns that share a name between the two dataframe
pd.merge(df1, df2, on = ['id','Col_4'])  # Multiple condtions specified in a list
```

```python
pd.merge(df1, df2, suffixes =['_l','_r'], left_on='Col_2', right_on='Col_A') # this is way data frame can be merged with different names
```

```python
pd.merge(df1, df2, suffixes =['_left','_right'], left_on='Col_2', right_on='Col_A') # excercise
```

### Joining on indices

```python
pd.merge(df1, df2, suffixes =['_l','_r'], left_index=True, right_index=True)
```

### Join Types in Pandas

 - Inner Join : pd.merge(df1, df2, on ="col_name") ---> its default, Join by common elements betweeen sets
 - Outer Join : pd.merge(df1, df2, on ="col_name", how="outer") --> combine all between sets (common + uncommon)/ Full Join
 - Left Join : pd.merge(df1, df2, on ="col_name", how="left") --> Join common elements between sets + all from left table/ set
 - Right Join : pd.merge(df1, df2, on ="col_name", how="right") --> Join common elements between sets + all from right table / set
 - We can specify the type of join by 'how' arguments

```python
df1
```

```python
df2
```

#### Outer Join

```python
pd.merge(df1, df2, on='Col_4', how='outer', suffixes=['_l','_r'])
```

#### Left Join

```python
pd.merge(df1, df2, on='Col_4', how='left', suffixes=['_l',['_r']])
```

#### Right Join

```python
pd.merge(df1, df2, on='Col_4', how='right', suffixes=['_l',['_r']])
```

# Concat

- Good for side-by-side & stacking DataFrame on top of each other
- More details https://realpython.com/pandas-merge-join-and-concat/
- Merge used for combining dataset side by side, while concat comining dataset on top of each other
- syntax : pd.concat & pass df as list 
- Concat has not no " left & Right" Join

```python
df1
```

```python
df2
```

```python
# Default axis is 0 (this stacks the DataFrames)
pd.concat([df1, df2])
```

```python
# Reset the index
pd.concat([df1, df2], ignore_index= True)  # When at index we dont want to have repeated value, use ignore_index= True
```

```python
# Can side-side-side by specifying axis=1 / column wise
pd.concat([df1, df2], axis=1)
```

```python
# default join is outer, but you can specify inner join
# there are no option for left or right joins
pd.concat([df1, df2], axis=1, join='inner')
```

```python
pd.concat([df1, df2], axis=0, join='inner')  
```
