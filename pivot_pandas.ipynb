### About dataset2

# Pivot Table

- Pivot table is used to summazie & aggregate data inside dataframe

```python
df2.pivot_table(index='city', columns='date')  # got the average of city (e.g (85+26/2=55.5 for mumbai on 5 Feb 21) )
```

```python
df2.pivot_table(index='city', columns='date', aggfunc="mean")  # mean is default
```

```python
df2.pivot_table(index='city', columns='date', aggfunc="sum")  # Applying aggreation on data frame by 'aggfunc'
```

```python
df2.pivot_table(index='city', columns='date', aggfunc="count")  # Count Application
```

```python
df2.pivot_table(index='city', columns='date', margins=True)  # marging will all to add all (average) as new col found as "All", default false
```

# Grouper

- fromt the data table 3, we can aggregate based on date frequency
- lets say you want to know the average temperature for the month of May & average temp for month of December
- This is something can be made by pivot function grouper
- Doc: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases

#### Grouper Arguments

 - freq: M - month end frequency

# Load Data - 3

```python
df3 = pd.read_csv('weather3.csv')
df3
```

```python
# df3.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city') # will return type error, because date needed to be convered first
```

```python
df3['date']=pd.to_datetime(df3['date'])  # Str date column converted into real datetime
df3
```

```python
type(df3['date'][0]) # Verification, 0 mean 1st index/ date columns, yes its Timestamp
```

```python
# Based on date got average temp for the month of May & Dec, it will show end date of May & end data of Dec
df3.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city') 
```

```python
df3.pivot_table(index=pd.Grouper(freq='W', key='date'), columns='city') # Freq: w- Weekly
```

```python
df3.pivot_table(index=pd.Grouper(freq='BH', key='date'), columns='city') # Frea: BH- Business Hour Freq
```

```python
df3.pivot_table(index=pd.Grouper(freq='QS', key='date'), columns='city') # Freq: Quarter Start Freq
```

```python
df3.pivot_table(index=pd.Grouper(freq='B', key='date'), columns='city')  # Freq:B- Business Day Freq
```
