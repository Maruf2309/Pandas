### Normalizing

```python
m = df.SepalLengthCm.mean()    # Mean of SepalLength
std = df.SepalLengthCm.std()   # Std of SepalLength
df.SepalLengthCm.apply(lambda val: val-m/std)  # Normalizing the value
```

# Filtering

- As like Numpy we can filter here too

```python
df['SepalLengthCm'] > 5
```

### And

```python
df[(df['SepalLengthCm']>5) & (df['PetalLengthCm']<2)]   # Multiple conditions with bitwise operator & not and
```

### Or

```python
df[(df['SepalLengthCm']>5) | (df['PetalLengthCm']<2)]   # Or Function
```

### NAGATION

- Nagation sign used for returning all except True Condition or return all false statement

```python
df[~(df['SepalLengthCm']>5)]  # Return SepalLength value when < 5
```

```python
df[~(df['SepalLengthCm']>5)] ['Species']  # Finding Species in respect of SepalLength condition - Single Col
```

```python
df[~(df['SepalLengthCm']>5)] [['PetalWidthCm','Species']]  # Multiple col with respect to SepalLenght Condtions
```

# Groupby

- Group by will be applied for aggration based on Category
- By : Category

```python
df.groupby(by='Species').aggregate(np.median)     # Median by Category
```

```python
df.groupby(by='Species').aggregate(np.mean)  # Mean by Category
```

```python
df.groupby(by='Species').aggregate(np.sum) # Sum by Category
```
