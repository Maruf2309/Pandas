# Alternaive Ways to Merge Multiple Files

```python
from glob import glob  # Glob is a function for merge multiple files for more efficiently or efficient file listing
```

```python
files = glob('sales*')   # the word 'sales' is the first part of file name
files
```

```python
files.sort()  # Files Shorted
files
```

```python
data2 = pd.concat((pd.read_csv(file)for file in files), ignore_index=True)  # One line code for concatination
```

```python
data2.head(5000)
```

# Exporting Data to CSV

```python
data2.to_csv("data2.csv", index=False)
```
