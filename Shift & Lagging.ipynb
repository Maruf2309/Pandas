# Shifting & Lagging

```python
import pandas as pd
```

```python
df = pd.read_csv('stock_price.csv', parse_dates=['Date'], index_col='Date')  # parse_date to convert str(date) to datetime
df
```

```python
df.drop(df.columns[df.columns.str.contains('Unnamed: 2',case = False)],axis = 1, inplace = True) # Remove col "Unnamed:2"
```

```python
df.drop(df.columns[df.columns.str.contains('Unnamed: 3',case = False)],axis = 1, inplace = True) # Remove col "Unnamed:3"
```

- Shift method works  either on dataframe or datetime

```python
df.shift(1)
```

```python
df.shift(2)
```

### Addng New columns and show last day price

```python
df['Previous Day Price'] = df['Price'].shift(1)
df
```

```python
df['1 Day Price Change'] = df['Price'] - df['Previous Day Price']
df
```

### Return Percentage

```python
df['2 Day % Return'] = (df['Price'] - df['Price'].shift(2)) *100 / df['Price'].shift(2)
df
```

```python
df = df[['Price']]  # Get only Price col
df
```

```python
df.index  # Index Freq = None
```

```python
df.index = pd.date_range(start='2022-12-05', periods=16, freq='B')  # B: Business Frequency, period cant be <16, since 16 rows
df.index
```

```python
df  # Prious 9th day is missing, it comes now due to setting Buiness Frequency, B
```

```python
df.tshift(1)  # tshift adjust the date not, values, earlier start date was 5 th Dec, now its 6th Dec
```
